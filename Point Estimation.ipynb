# Point Estimation

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

## Normal Probablity Distribution

```python
p = 0.05
x = np.linspace(-4, 4, 100) # plot x range
y = stats.norm.pdf(x) # plot y : probability density function
ep = stats.norm.ppf(p) # estimate point
ax = np.linspace(-4, ep, 100) # x range fill area
ay = stats.norm.pdf(ax) # y range of fill area
r = stats.norm.rvs(size=10000) # generate random values

# Draw Plot
plt.figure(figsize=(15,5))
plt.title('Normal Probablity Distribution')
plt.hist(r, 100, density=True, histtype='stepfilled', alpha=0.2)
plt.plot(x, y, 'b-')
plt.axvline(x=stats.norm.ppf(p), color='r')
plt.axvline(x=stats.norm.ppf(1-p), color='r')
#plt.fill_between(ax, 0, ay, color='r', alpha=0.6)
plt.show()

print('Probablity:{0}, x:{1}'.format(p, stats.norm.ppf(p)))
print('Probablity:{0}, x:{1}'.format(1-p, stats.norm.ppf(1-p)))
```
