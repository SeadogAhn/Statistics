# Basic Properties of Confidence Intervals

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

p = 0.95
x = np.linspace(-4., 4., 100) # plot x range
y = stats.norm.pdf(x) # plot y : probability density function
```

### Greater than or Less than

```python
ep = stats.norm.ppf(p) # estimate point
ax = np.linspace(-4, ep, 100) # x range fill area
ay = stats.norm.pdf(ax) # y range of fill area
print('Probablity:{0}, x:{1}'.format(p, stats.norm.ppf(p)))
plt.figure(figsize=(15,5))
plt.plot(x, y, 'b-')
plt.axvline(x=stats.norm.ppf(p), color='r')
plt.fill_between(ax, 0, ay, color='b', alpha=0.6)
plt.show()
```

### Not equal

```python
ep1 = stats.norm.ppf((1-p)/2) # estimate point
ep2 = stats.norm.ppf(1-(1-p)/2) # estimate point
ax = np.linspace(ep1, ep2, 100) # x range fill area
ay = stats.norm.pdf(ax) # y range of fill area
print('Probablity:{0}, -x:{1}, x:{2}'.format(p, stats.norm.ppf((1-p)/2), stats.norm.ppf(1-(1-p)/2)))
plt.figure(figsize=(15,5))
plt.plot(x, y, 'b-')
plt.axvline(x=stats.norm.ppf((1-p)/2), color='r')
plt.axvline(x=stats.norm.ppf(1-(1-p)/2), color='r')
plt.fill_between(ax, 0, ay, color='b', alpha=0.6)
plt.show()
```
